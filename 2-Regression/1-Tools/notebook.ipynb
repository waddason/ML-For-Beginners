```python
# Print a portion of the data
# select column 6 = blood sugar levelody mass index
def print_data(X_full, column_index, column_name):
    X = X_full[:, column_index]
    X = X.reshape((-1, 1))
    
    # Split the data into training/testing sets
    X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)

    # Train the model
    model = linear_model.LinearRegression()
    model.fit(X_train, y_train)

    # Test the model
    y_pred = model.predict(X_test)

    # Show the data
    plt.scatter(X_test, y_test,  color='black')
    plt.plot(X_test, y_pred, color='blue', linewidth=3)
    plt.xlabel('Scaled ' + column_name)
    plt.ylabel('Disease Progression')
    plt.title('A Graph Plot Showing Diabetes Progression Against' + column_name)
    plt.show()
```

```python
print_data(X_full, 4, 'total serum cholesterol')
```

```python
print_data(X_full, 5, 'low density lipoproteins')
```

```python
print_data(X_full, 6, 'High-density lipoproteins')
```

```python
print_data(X_full, 7, 'Total cholesterol')
```

```python
print_data(X_full, 8, 'Possibly log of serum triglycerides level')
```

```python
print_data(X_full, 9, 'Blood Sugar level')
```
